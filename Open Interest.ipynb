```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
```

```python
ticker = yf.Ticker("SPY")
option_chain = ticker.option_chain()
option_chain
```

```python
# extracting calls and puts option data
calls = option_chain.calls
puts = option_chain.puts
```

```python
historical_data = ticker.history(period="1d")
current_stock_price = historical_data["Close"].iloc[-1]
```
