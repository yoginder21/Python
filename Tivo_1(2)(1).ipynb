Q7 7.	Based on your analysis of the dataset, who do you believe the ideal customer group  who marketing should focus their attention on? 

```python
average_income = df['Annual Income (x1000 $)'].mean()
```

```python
filtered_df_analysis = df[(df['Annual Income (x1000 $)'] > average_income) &
                 (df['Purchasing Frequency (every x months)'] > 12) &
                 (df['Technology Adoption'] == 'early') &
                 (df['Age'] < 50)]
```

```python
filtered_df_analysis
```

To focus on a specific group of customers based on the criteria  mentioned above,  we can do so using data filtering techniques to get more business. Here's a general approach to achieve this:

Filter by Income: Identify customers whose income is greater than the average income of all customers.

Filter by Purchase Frequency: Customers who have a purchase frequency of more than 12.

Filter by Technology Adoption: Early adopters of technology.

Filter by Age: Finally, from this group, we will select only those customers who are less than 50 years old.
